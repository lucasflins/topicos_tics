```python
from datasets import load_dataset
import pandas as pd
import numpy as np

# Load the MS-MARCO dataset
dataset = load_dataset('ms_marco', 'v2.1', split='train',)
```

```python
data = dataset.select(np.random.randint(0,dataset.shape[0],10_000)).to_pandas()
```

```python
data = data[['query', 'query_type']]

# Filter out rows with missing query_type
data = data.dropna(subset=['query_type'])
```

```python
query_type_map = {query_type: idx for idx, query_type in enumerate(data['query_type'].unique())}
data['label'] = data['query_type'].map(query_type_map)
```

```python
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments,EarlyStoppingCallback
import torch
from sklearn.model_selection import train_test_split
import numpy as np

# Preprocess data for classification
train_texts, val_texts, train_labels, val_labels = train_test_split(data['query'].tolist(), data['label'].tolist(), test_size=0.2)

# Load tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased',model_max_length=32)
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=len(query_type_map))
model = model.to('cuda')

# Tokenize data
train_encodings = tokenizer(train_texts, truncation=True, padding=True,)
val_encodings = tokenizer(val_texts, truncation=True, padding=True)

class MSMarcoDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = MSMarcoDataset(train_encodings, train_labels)
val_dataset = MSMarcoDataset(val_encodings, val_labels)

# Training arguments
training_args = TrainingArguments(
    output_dir='.cache',
    num_train_epochs=10,
    evaluation_strategy='epoch',
    save_strategy='epoch',
    logging_strategy='epoch',
    per_device_train_batch_size=128*3,
    per_device_eval_batch_size=64,
    warmup_ratio=0.1,
    learning_rate=1e-5,
    # tf32=True,
    weight_decay=0.01,
    logging_dir='.cache/logs',
    load_best_model_at_end = True,
    max_grad_norm=5.0,
    save_total_limit=1,
    log_level='info',
    # fp16=True,
)

trainer = Trainer(
    model=model,
    tokenizer=tokenizer,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=2)],
)

# Train the model
trainer.train()

# Save the model and tokenizer
trainer.save_model('/content/drive/MyDrive/mestrado_tics/final_model')
# tokenizer.save_pretrained('final_model')
```

```python
from tqdm import tqdm

def get_embeddings(texts, model, tokenizer, batch_size=32):
    embeddings = []

    for i in tqdm(range(0, len(texts), batch_size)):
        batch_texts = texts[i:i + batch_size]
        encodings = tokenizer(batch_texts, truncation=True, padding=True, return_tensors='pt')
        encodings = encodings.to('cuda')
        with torch.no_grad():
            outputs = model.bert(**encodings)
            batch_embeddings = outputs.last_hidden_state[:, 0, :].detach().cpu().numpy()
            embeddings.append(batch_embeddings)

    embeddings = np.vstack(embeddings)
    return embeddings

# Ensure the model is in evaluation mode and moved to GPU
model.eval()
model.to('cuda')

# Generate and save embeddings for the entire dataset
embeddings = get_embeddings(data['query'].tolist(), model, tokenizer)
np.save('/content/drive/MyDrive/mestrado_tics/embeddings.npy', embeddings)
```

```python
import hdbscan
# Perform clustering
clusterer = hdbscan.HDBSCAN()
clusters = clusterer.fit_predict(embeddings)

# Add clusters to dataframe
data['cluster'] = clusters
```

```python
model.to('cuda')
model.eval()
def batch_predict(texts, model, tokenizer, batch_size=32):
    predictions = []

    for i in tqdm(range(0, len(texts), batch_size)):
        batch_texts = texts[i:i + batch_size]
        encodings = tokenizer(batch_texts, truncation=True, padding=True, return_tensors='pt')
        encodings = encodings.to('cuda')

        with torch.no_grad():
            outputs = model(**encodings)
            logits = outputs.logits
            batch_predictions = torch.argmax(logits, dim=1).cpu().numpy()
            predictions.extend(batch_predictions)

    return predictions
```

```python
data['pred'] =  batch_predict(data['query'].tolist(), model, tokenizer,batch_size=8)
```
